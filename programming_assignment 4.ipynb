# Problem
There are several problems with this dataset, one of them being that a few of the columns have missing data entries where 3 of the columns in the dataset has the whole rows completely empty. By using our quantDDA function in assignment 2, which is a more in depth description of our data set compared to the pandas describe function(), we will take a look at the dataset.

```python
# For this assignment, need to display:
# 1.Situation 
# 2.Problem 
# 3.Solutions 
# 4.Next steps 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('SAFI_results.csv')
pd.set_option('display.max_columns', 100)

def quantDDA(data):
    unique_entries = data.nunique() 
    missing_entries = data.isnull().sum()
    mode = data.mode()[0:1] 
    mean = data.mean()
    std = data.std()
    minim = data.min()
    maxim = data.max()
    q3 = data.quantile(0.75)
    q2_median = data.quantile(0.5)
    q1 = data.quantile(0.25)
    IQR = q3 - q1
    extreme_vals_lower = q3 - (3 * IQR)
    extreme_vals_upper = q3 + (3 * IQR)
    outliers = ((data < (q1 - 1.5 * IQR)) | (data > (q3 + 1.5 * IQR))).sum()
    skew = data.skew()
    kurtosis = data.kurt()
    
    
    #Part 3
    data["MissingValueCount"] = data.apply(lambda x: x.isnull().sum(), axis=1) #creating new column for dataset
    display(data)
    
    print('Number of observations:', len(data))
    print('Number of entries:', data.size,'\n\n')
    print('Displaying mode:')
    display(mode)
    
    result = pd.concat([
                      unique_entries.rename('unique_entries'),  
                      missing_entries.rename('missing_entries'),
                      extreme_vals_lower.rename('extr_val_lower'),
                      extreme_vals_upper.rename('extr_val_upper'),
                      outliers.rename('outliers'),
                      mean.rename('mean'),
                      std.rename('std_dev'),
                      maxim.rename('max'),
                      minim.rename('min'),
                      q3.rename('Q3'),
                      q2_median.rename('Q2 (med)'),
                      q1.rename('Q1'),
                      skew.rename('skew'),
                      kurtosis.rename('kurtosis')], axis=1).T #wasn't sure if we keep the 1st column unamed so I left it there incase
    print('\n\n Displaying rest of the quantitative desciptions:')
    display(result)
    
from IPython.display import HTML

HTML('''<script>
code_show=true; 
function code_toggle() {
 if (code_show){
 $('div.input').hide();
 } else {
 $('div.input').show();
 }
 code_show = !code_show
} 
$( document ).ready(code_toggle);
</script>
<form action="javascript:code_toggle()"><input type="submit" value="Click here to toggle on/off the raw code."></form>''')
```

```python
quantDDA(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def vizDDA(data):
    sns.pairplot(data)

vizDDA(df)
```

With the help of seaborn pairplot, we can also notice that we get empty graphs which won't be of any use.

Having empty values in our dataset is a major problem and it is very common to encounter missing entries in a dataset. Missing data present various problems such as follows:
- The absence of data reduces statistical power.
- The lost data can cause bias in the estimation of parameters. 
- It can reduce the representativeness of the samples. 
- It may complicate the analysis of the study. 
- Reduce performance of data mining algorithms

Each of these distortions may threaten the validity of the trials and can lead to invalid conclusions.

# Solutions
In order to solve this problem, several steps can be taken:
- For categorical values, we can flag them with "Missing".
- For numerical values, we can flag them with 0.
- For continuous variables, we can flag them with a featre called "Missing V".
- For a column with too many missing values, we can drop the column.
- For observations with many missing values, we can drop the observation or fix them.


There is no right way of dealing with missing values, and it will all change the way we make our analysis and making close to accurate predictions. For the problem we are dealing with, since the 4 columns mentionned above have 131 missing columns, we will be dropping the the columns.

```python
to_drop = ['C02_respondent_wall_type_other', 'F05_money_source_other', 'F10_liv_owned_other', '_note']

df.drop(to_drop, inplace=True, axis=1)
```

```python
quantDDA(df)
```

```python
vizDDA(df)
```

We can now see that there are no empty graphs in our pair plot which makes it more readable and we can see better relationships between the variables. If the dataset was even larger, having so many empty graphs in our pairplot will make it harder to see the rest of the graphs and will take unecessary space.
